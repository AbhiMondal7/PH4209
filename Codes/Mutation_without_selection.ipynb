```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# parameters
# two type A and B

x = 0.3 # type A
u1 = 0.02
y = 1-x # type B
u2 = 0.95

n = 20 #population size
time = 100 # number of generation
```

```python
def isMutation(u):
    rN = np.random.rand();
    if rN > u:
        return 0
    else:
        return 1
```

```python
# z = frequency
# u = mutation prob
# n = pop size
def newMut(z,u,n):
    z_n = z*n
    new_mutation = 0
    for i in range(int(z_n)):
        if isMutation(u):
            new_mutation+=1
    return new_mutation
```

```python
# Time evolution
def generation(time,x,y,u1,u2,n):
    x_arr = np.zeros(time+1,dtype=float)
    y_arr = np.zeros(time+1,dtype=float)
    x_arr[0] = x
    y_arr[0] = y
    for t in range(1,time+1):
        y_new = newMut(x,u1,n)
        x_new = newMut(y,u2,n)
        x_pop = (x*n) + x_new - y_new
        y_pop = (y*n) + y_new - x_new
        x = x_pop / n
        y = y_pop / n 
        x_arr[t] = x
        y_arr[t] = y
    return x_arr, y_arr
```

```python
# plotting 
x_arr, y_arr = generation(time,x,y,u1,u2,n)
t_arr = np.arange(time+1)
x_star = u2/(u1+u2)
y_star = u1/(u1+u2)
fig, ax = plt.subplots(figsize = (8,6))

ax.plot(t_arr,x_arr, marker = 'o', label = "type A")
ax.plot(t_arr,y_arr, marker = 'o', label = "type B")
plt.axhline(y=x_star, color='r', linestyle='--', linewidth=2, label='Threshold for type A')
plt.axhline(y=y_star, color='g', linestyle='--', linewidth=2, label='Threshold for type B')
ax.set_xlabel('Generation')
ax.set_ylabel('Frequency')
ax.set_title('Freq over generation')
ax.legend(loc = 'best')
ax.grid(True)
plt.show()
```
