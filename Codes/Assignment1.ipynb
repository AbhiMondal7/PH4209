# Evolution via mutation without selection

```python
# pakages 
import numpy as np
import matplotlib.pyplot as plt
import random
```

## Part 1

```python
# parameters
N_values = np.array([50,1000,10000]) # population size
T = 2000 # number of generation

u1 = 0.003 # mutation rate for type 0
u2 = 0.001 # mutation rate for type 
```

```python
freq0  = np.zeros((len(N_values),T))
freq1 = np.zeros((len(N_values),T))
```

```python
for i,N in enumerate(N_values):
    # initialise population
    population = np.array([0]*(N//2) + [1]*(N//2))
    np.random.shuffle(population)
    for t in range(T):
        pop0 = (population == 0) # find the type 0
        rand0 = np.random.rand(N) 
        population[pop0 & (rand0 < u1)] = 1 # mutations 0 -> 1
    
        pop1 = (population == 1)
        rand1 = np.random.rand(N)
        population[pop1 & (rand1 < u2)] = 0 #mutation 1-> 0
    
        freq0[i,t] = np.mean(population == 0)
        freq1[i,t] = np.mean(population == 1)
```

```python
# plotting
fig, axs = plt.subplots(2, 2,figsize = (13,11))

gen_arr = np.arange(1,T+1)
x_star = u2/(u1+u2) #ideal type 0
y_star = u1/(u1+u2) # ideal type 1

axs = axs.flatten()
for i,N in enumerate(N_values):
    axs[i].scatter(gen_arr,freq0[i,:],label = f"type0",s = 3)
    axs[i].scatter(gen_arr,freq1[i,:],label = f"type1",s = 3)
    axs[i].set_title(f"N = {N}")
    axs[i].axhline(y=x_star, color='r', linestyle='--', linewidth=2, label='Threshold for type 0')
    axs[i].axhline(y=y_star, color='g', linestyle='--', linewidth=2, label='Threshold for type 1')
    axs[i].set_xlabel("Generation")
    axs[i].set_ylabel("Frequency")
    axs[i].legend(loc = "best")
for i,N in enumerate(N_values):
    axs[3].scatter(gen_arr,freq0[i,:],label = f"type0, N = {N}",s = 3)
    axs[3].scatter(gen_arr,freq1[i,:],label = f"type1, N = {N}",s = 3)
    axs[3].set_xlabel("Generation")
    axs[3].set_ylabel("Frequency")
    axs[3].set_title("Comparison over the population size")

axs[3].axhline(y=x_star, color='r', linestyle='--', linewidth=2, label='Threshold for type 0')
axs[3].axhline(y=y_star, color='g', linestyle='--', linewidth=2, label='Threshold for type 1')
axs[3].legend(loc = "best")
plt.show()
```
